### Задача № 4 ##

Даны две группы студентов. В каждой по 21 человеку. Первая группа прошла недельную
подготовку для сдачи тестов, другая группа приступила к тестам без подготовки.
В 1й группе среднее время прохождения теста целиком требует 1.8 ч, во второй 2.3 ч.
Постройте 95% интервал для разности средних арифметических, если известно, что
стандартное отклонение 1 группы составило 0.4 ч, во 2й группе -0.6ч

```python
n_1 = n_2 = 21
X_1 = 1.8
X_2 = 2.3
D_1 = 0.4
D_2 = 0.6
```

Пусть у нас будет 2 выборки по 21 человеку. Если мы измерим по ним средние арифметические, то получим ▁X_1  и   ▁X_2 и их разность Δ – будет оценкой истинной разницы. Дисперсии в каждой группе различны. Поэтому для начала найдем общую дисперсию, как среднюю дисперсию обеих групп.

D=1/2(D_1+D_2), где   D_1 и D_2 – дисперсии обеих групп

```python
D= 1/2 * (D_1**2+D_2**2) # обьединённая оценка дисперсии
D
```

А вместо стандартной ошибки среднего  σ/√n найдем стандартную ошибку разности средних: S_Δ=√(D/n_1 +D/n_2 ), где n_1 и  n_2 – объемы выборок.

```python
S_Δ = np.sqrt(D/n_1 + D/n_2)
S_Δ
```

И таким образом, формула для нахождения доверительного интервала для разности средних арифметических µ_1-µ_2 будет:
Δ± t_(α/2)* S_Δ , где Δ= X_1 - X_2


В данном случае для критерия Стьюдента степени свободы df = n1 + n2 - 2, где n1 и n2 - размеры выборок

```python
df = n_1 + n_2 - 2
df
```

```python
# найдём разность средних
delta = X_1 - X_2
delta
```

```python
SE = np.sqrt(D/n_1 + D/n_2) # стандартная ошибка разности средних
SE
```

```python
import scipy.stats as stats
```

```python
t = stats.t.ppf(0.975, 40)
t
```

Теперь осталось посчитать верхнюю и нижнюю границы доверительного интервала L и  U соответственно.

```python
L = delta - t * SE
L
```

```python
U = delta + t * SE
U
```

```python
print("Доверительный интервал : [", round(L, 2), ";", round(U, 2), "]")
```
